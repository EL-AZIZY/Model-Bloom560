```python
import transformers
from transformers import BloomForCausalLM
from transformers import BloomTokenizerFast
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
bloom_560m= BloomForCausalLM.from_pretrained("bigscience/bloom-560m")
```

```python
tokenizer= BloomTokenizerFast.from_pretrained("bigscience/bloom-560m")
```

```python
prompt1 = "creer moi un code python pour calculer la somme de 1 a 100"
resultat_length= 128
inputs1=tokenizer(prompt1,return_tensors="pt")
```

```python
resultat= (tokenizer.decode(bloom_560m.generate(inputs1["input_ids"],
                                                max_length= resultat_length,
                                                do_sample=True,
                                                top_k=50,
                                                top_p= 0.9
                                                )[0]))
print("Q:" ,prompt1)
print("completion: " ,resultat)
```
